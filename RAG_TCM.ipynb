## Self-defined Class and Functions

```python
# 準備RAG系統
class MyWebLoader(WebBaseLoader):
    def _scrape(self, url: str, **kwargs) -> BeautifulSoup:
        context = ssl.create_default_context()
        context.check_hostname = False
        context.verify_mode = ssl.CERT_NONE
        kwargs={'verify': False}
        response = requests.get(url, **kwargs)
        response.raise_for_status()
        return BeautifulSoup(response.text, "html.parser")
    
    def load(self) -> List[Document]:
        docs = []
        for path in self.web_paths:
            print(path)
            soup = self._scrape(path, bs_kwargs=self.bs_kwargs)
            text = soup.text
            metadata = {"source": path}
            docs += [Document(page_content=text, metadata=metadata)]
        return docs
    
def create_or_load_vectorstore(urls, document_embedder, vector_store_path = "vectorstore.pkl"):
    if os.path.exists(vector_store_path):
        with open(vector_store_path, "rb") as f:
            return pickle.load(f)
    else:
        loader = MyWebLoader(urls)
        raw_documents = loader.load()
        if raw_documents:            
            text_splitter = RecursiveCharacterTextSplitter(
                chunk_size=500,
                chunk_overlap=100,
                separators=["\n\n", "\n", ".", ";", ",", " ", ""],
            )
            documents = text_splitter.split_documents(raw_documents)
            
            print("Adding document chunks to vector database...")
            vectorstore = FAISS.from_documents(documents, document_embedder)
            print("Saving vector store")
            try:
                with open(vector_store_path, "wb") as f:
                    pickle.dump(vectorstore, f)
                print("Vector store created and saved.")
            except:
                print("No documents available to process!")
            return vectorstore
        return None
    
# 解析評估結果    
def parse_result(result, criteria):
    parsed_result = {}
    
    # 取得解釋
    explanation_match = re.search(r'Explanation:(.*?)Rating:', result['reasoning'], re.DOTALL)
    if explanation_match:
        parsed_result['explanation'] = explanation_match.group(1).strip()
    else:
        parsed_result['explanation'] = None
    
    # 取得各個 criteria 的分數
    for criterion in criteria:
        score_match = re.search(rf'{criterion.capitalize()}:\s*\[\[(\d+(?:\.\d+)?)\]\]', result['reasoning'])
        if score_match:
            parsed_result[criterion] = float(score_match.group(1))
        else:
            parsed_result[criterion] = 0
    
    # # 總體評分
    # overall_match = re.search(r'Overall rating:\s*\[\[(\d+(?:\.\d+)?)\]\]', result['reasoning'])
    # if overall_match:        
    #     parsed_result['overall_score'] = float(overall_match.group(1))
    # else:
    #     parsed_result['overall_score'] = result.get('score')
    
    return parsed_result

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

def get_sources(docs):
    return "\n".join(set(doc.metadata["source"] for doc in docs))
```

```python
# 準備測試數據
test_data = [
    {"question": "中醫癌症術後治療及調理門診的內容為何?", 
     "answer": "協助癌症患者西醫療程的順利完成，並針對癌症患者的惡病質，如體重減輕、倦怠、食欲不振、疼痛等，透過中醫的體質調理，以提升更好的生活品質及治療效果。"},
    {"question": "中醫脊椎門診的內容為何?", 
     "answer": "神經骨骼肌肉系統的病症，可透過整脊增加關節活動度，調整錯位的脊椎對身體的傷害，是必要且重要的治療，對於有整脊需求之患者，本院特開設健康脊椎特別門診。"},
    {"question": "中醫體重控制諮詢的內容為何?", 
     "answer": "體重過重對健康的威脅很大，包括高血脂、高血壓、糖尿病、心血管疾病的發生率大幅增加，本院區針對有體重控制困擾及疑問的民眾，提供最完整的諮詢服務以及治療，包括中醫針灸、耳穴、拔罐、中藥藥粉或中藥即飲包、飲食及運動諮詢等。"},
    # {"question": "", "answer": ""}
    # 添加更多測試問題...
]
```

```python
# Defined LLM
llm = ChatNVIDIA(
    model="meta/llama3-70b-instruct",
    api_key=os.environ["NVIDIA_API_KEY"],
    temperature=0.1,
    max_tokens=1024,
)

# How many retrieved documents to keep at each step 
top_k_documents_retriever = 30
top_n_documents_reranker = 5

embedding_model = NVIDIAEmbeddings(
    model="nvidia/nv-embedqa-e5-v5", 
    api_key=os.environ["NVIDIA_API_KEY"], 
    truncate="NONE", 
)

urls = ["https://tpech.gov.taipei/Content_List.aspx?n=CABD5B0B973F4534", #聯醫 中醫
        ### 可以多個網站 ###
       ]
vectorstore = create_or_load_vectorstore(urls, embedding_model)
retriever = vectorstore.as_retriever(search_kwargs={'k': top_k_documents_retriever}) 



reranker = NVIDIARerank(#model="nvidia/nv-rerankqa-mistral-4b-v3",
                        top_n=top_n_documents_reranker,
                        tuncate="NONE")
#################################################################
pure_llm_template = """
你是聯合醫院的中醫助理。請根據上下文信息來提供準確的中文回答。：

問題：{question}

回答："""

pure_llm_prompt = PromptTemplate(
    input_variables=["question"],
    template=pure_llm_template
)

pure_llm_chain = pure_llm_prompt | llm


rag_template = """
你是聯合醫院的中醫助理。請根據上下文信息來提供準確的中文回答。
如果無法從上下文中找到答案，請說明你無法確定，並建議諮詢醫院的專業醫師。

上下文：{context}

問題：{question}

回答：請提供詳細答案，並在回答末尾列出使用的參考來源。
參考來源：{sources}"""

rag_prompt = PromptTemplate(
    input_variables=["context", "question", "sources"],
    template=rag_template
)

retriever = ContextualCompressionRetriever(
    base_compressor=reranker,
    base_retriever=retriever
)
#################################################################


rag_chain = (
    {"context": retriever | format_docs, 
     "question": RunnablePassthrough(), 
     "sources": retriever | get_sources} 
    | rag_prompt 
    | llm
)



# 設置criteria
criteria = {
    "accuracy": "評估回答的準確性，1分表示完全不準確，10分表示非常準確。",
    "completeness": "評估回答的完整性，1分表示非常不完整，10分表示非常完整。",
    "harmfulness": "評估回答的傷害性，1分表示非常不佳，10分表示非常良好。",
}
qa_evaluator = load_evaluator("labeled_score_string", criteria=criteria, llm=llm)



# Inference
results = []
for item in test_data:
    question = item["question"]
    reference_answer = item.get("answer")

    pure_llm_answer = pure_llm_chain.invoke({"question": question})
    rag_answer = rag_chain.invoke(question)

    if reference_answer:
        pure_llm_qa_result = qa_evaluator.evaluate_strings(
            prediction=pure_llm_answer.content,
            reference=reference_answer,
            input=question
        )
        rag_qa_result = qa_evaluator.evaluate_strings(
            prediction=rag_answer.content,
            reference=reference_answer,
            input=question
        )

        parsed_pure_llm_result = parse_result(pure_llm_qa_result, criteria.keys())
        parsed_rag_result = parse_result(rag_qa_result, criteria.keys())

        result = {
            "Question": question,
            "Reference Answer": reference_answer,
            "Pure LLM Answer": pure_llm_answer.content,
            "Pure LLM Result": parsed_pure_llm_result,
            "RAG Answer": rag_answer.content,
            "RAG Result": parsed_rag_result,
        }
        results.append(result)

# 將結果轉換為DataFrame
df_results = pd.DataFrame(results)

# 結果輸出
for index, row in df_results.iterrows():
    print(f"\n問題 {index + 1}: {row['Question']}")
    print(f"答案 {index + 1}: {row['Reference Answer']}")
    print(colored("\n純LLM回答:", "blue"))
    print(colored(row['Pure LLM Answer'], "blue"))
    print(colored("純LLM評估結果:", "red"))
    for criterion, score in row['Pure LLM Result'].items():
        if criterion != 'explanation':
            print(f"{criterion.capitalize()}: {score:.2f}")
    print(colored(f"解釋: {row['Pure LLM Result']['explanation']}", "red"))
    # print(f"解釋: {row['Pure LLM Result']['explanation']}")

    print(colored("\nRAG回答:", "green"))
    print(colored(row['RAG Answer'], "green"))
    print(colored("RAG評估結果:", "red"))
    for criterion, score in row['RAG Result'].items():
        if criterion != 'explanation':
            print(f"{criterion.capitalize()}: {score:.2f}")
    # print(f"解釋: {row['RAG Result']['explanation']}")
    print(colored(f"解釋: {row['RAG Result']['explanation']}", "red"))

    print("\n" + "="*50)

# 計算平均分數
average_scores = {criterion: {} for criterion in criteria.keys()}

for criterion in criteria.keys():
    pure_llm_avg = df_results['Pure LLM Result'].apply(lambda x: x[criterion]).mean()
    rag_avg = df_results['RAG Result'].apply(lambda x: x[criterion]).mean()
    average_scores[criterion]['Pure LLM'] = pure_llm_avg
    average_scores[criterion]['RAG'] = rag_avg

# 創建表格數據
table_data = []
for criterion, scores in average_scores.items():
    row = [criterion.capitalize()] + [f"{scores.get(model, 'N/A'):.2f}" if scores.get(model) is not None else 'N/A' for model in ['Pure LLM', 'RAG']]
    table_data.append(row)

# 表格輸出
print("\n總體平均分數:")
print(tabulate(table_data, headers=["評估標準", "Pure LLM", "RAG"], tablefmt="grid"))
```
